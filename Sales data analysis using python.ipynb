```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Sample Superstore data for python.csv",encoding='latin-1')
data.head()
```

```python
print(data)
```

```python
data.info()
```

```python
data[['Sales','Quantity','Discount','Profit']].describe()
```

```python
print(data.isnull().sum())
```

```python
print(data.isna().any(axis=1).sum())
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data.info()
```

```python
data.head()
```

```python
data['Order Month']= data['Order Date'].dt.month
data['Order Year']= data['Order Date'].dt.year
data.head()
```

## Monthly sales analysis

```python
sales_per_month=data.groupby('Order Month')['Sales'].sum().reset_index()
print(sales_per_month)
```

```python
months=range(1,13)
plt.bar(months,sales_per_month['Sales'])
plt.xticks(months)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales per Month')
plt.show()
```

## sales by category

```python
sales_per_category=data.groupby('Category')['Sales'].sum().reset_index()
print(sales_per_category)
```

```python
plt.pie(sales_per_category['Sales'],labels=sales_per_category['Category'],autopct='%1.1f%%')
plt.title('Sales per Category')
plt.legend(title="Category of products",loc="lower left")
plt.show()
```

## sales by sub-category

```python
sales_per_subcategory=data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=True).reset_index()
print(sales_per_subcategory)
```

```python
plt.barh(sales_per_subcategory['Sub-Category'],sales_per_subcategory['Sales'])
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.title('Sales per Sub-Category')
plt.show()
```

## Sales per Customer segment

```python
plt.bar(sales_per_customer_segment['Segment'],sales_per_customer_segment['Sales'])
plt.title("Sales per Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.show()
```

## City wise Sales

```python
city_sales=data.groupby('City')['Sales'].sum().sort_values(ascending=False).reset_index()
print(city_sales)
```

```python
threshold = 0.03  # 3%
city_sales['Percentage'] = city_sales['Sales'] / city_sales['Sales'].sum()

large = city_sales[city_sales['Percentage'] >= threshold]
small = city_sales[city_sales['Percentage'] < threshold]

others = small['Sales'].sum()

labels = list(large['City']) + ['Others']
sizes = list(large['Sales']) + [others]

plt.pie(sizes, labels=labels, autopct="%1.1f%%" , pctdistance=0.85)
plt.title("City sales across US")
plt.show()
```

## Most sold Products Name

```python
best_productname=data.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).reset_index().head(10)
print(best_productname)
```

```python
plt.barh(best_productname['Product Name'],best_productname['Quantity'],color="brown")
plt.title("Top 10 most sold products name")
plt.xlabel("Quantity")
plt.ylabel("Product name")
plt.xticks='Quantity'
plt.show()
```
